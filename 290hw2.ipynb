```python
#creation of the transition matrix:

#the lines represent the position and the columns are the posible movement 
#starting from this position

Transition_matrix=[[i+j for j in range(6)] for i in range(1,101)]
#the snake and ladder asides, we can move from 1 to 6 case upward 
#and it is equally likely

#we add then the snake and ladders:
Transition_matrix[0][0]=38
for i in range(len(Transition_matrix)):
    for j in range(len(Transition_matrix[i])):
        #the ladders:
        if Transition_matrix[i][j] == 4:
            Transition_matrix[i][j] = 14
        elif Transition_matrix[i][j] == 9:
            Transition_matrix[i][j] = 31
        elif Transition_matrix[i][j] == 28:
            Transition_matrix[i][j] = 84
        elif Transition_matrix[i][j] == 21:
            Transition_matrix[i][j] = 42
        elif Transition_matrix[i][j] == 36:
            Transition_matrix[i][j] = 44
        elif Transition_matrix[i][j] == 51:
            Transition_matrix[i][j] = 67
        elif Transition_matrix[i][j] == 71:
            Transition_matrix[i][j] = 91
        elif Transition_matrix[i][j] == 80:
            Transition_matrix[i][j] = 100
        #the snakes:
        elif Transition_matrix[i][j] == 16:
            Transition_matrix[i][j] = 6
        elif Transition_matrix[i][j] == 49:
            Transition_matrix[i][j] = 11
        elif Transition_matrix[i][j] == 47:
            Transition_matrix[i][j] = 26
        elif Transition_matrix[i][j] == 62:
            Transition_matrix[i][j] = 19
        elif Transition_matrix[i][j] == 64:
            Transition_matrix[i][j] = 60
        elif Transition_matrix[i][j] == 56:
            Transition_matrix[i][j] = 53
        elif Transition_matrix[i][j] == 87:
            Transition_matrix[i][j] = 24
        elif Transition_matrix[i][j] == 93:
            Transition_matrix[i][j] = 73
        elif Transition_matrix[i][j] == 95:
            Transition_matrix[i][j] = 75
        elif Transition_matrix[i][j] == 98:
            Transition_matrix[i][j] = 78
        elif Transition_matrix[i][j] > 100:
            Transition_matrix[i][j] = 100

        
```

```python
#Counting the different number of dice rolls on 5000 runs

import random

dict={}

for n in range(5000):
    dice_roll=0 
    position=0
    while position!=100:
        dice_roll+=1
        i=random.randint(0, 5)
        position=Transition_matrix[position][i]
    if dice_roll in dict:
        dict[dice_roll]+=1/5000
    else:
        dict[dice_roll]=1/5000
```

```python
#compute the distrubition of X

import matplotlib.pyplot as plt
X_max = max(dict.keys())
X=[i for i in range(X_max+1)]

X_distrib = [0] * (X_max + 1)

for key, valeur in dict.items():
    X_distrib[key] = valeur

plt.figure(figsize=(12, 4))
plt.plot(X,X_distrib)
plt.title("Distribution of X")
plt.xlabel("Number of dice rolls")
plt.ylabel("Probability")
plt.show()
```

```python
#compute the mean
import numpy as np

X_distrib_matrix=np.array(X_distrib)
X_matrix=np.array(X)

mean=np.dot(X_distrib_matrix,X_matrix.T)

mean
```
